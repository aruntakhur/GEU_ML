# Equation of the line is â€‹y = mx + c
1. Slope of the line is 1 i.e m = 1.
2. Intercept is 0 i.e. c = 0.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,10,11)
print(x)
```

```python
m = 1
y = m*x
plt.scatter(x,y)
plt.show()
```

```python
print(x)
print(y)
```

# Plotting Error function
It has only one paramter
LR_ErrorF.png

```python
w1 = np.linspace(-6,9,11)
error = 0
input = len(x)
for i in range(input):
  error = error + (w1*x[i]- y[i])**2
```

```python
print(error)
```

```python
plt.plot(w1,error)
plt.xlabel('W')
plt.ylabel('Error')
plt.title('W vs Error')
plt.scatter(w1,error,c='red')
plt.show()
```

From the above graph we can see that minimum value of Error is 0 when w is equivalent to 1
